```python
# Data Preprocessing:
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

# Make a copy to avoid modifying original data
df = df1.copy()

# 1. Impute missing values
# Age - fill with median
age_imputer = SimpleImputer(strategy='median')
df['Age'] = age_imputer.fit_transform(df[['Age']])
```

```python
# Drop Cabin due to high missing ratio
df1.drop(columns=['Cabin'], inplace=True)
df1
```

```python
label_encoders = {}
for col in ['Sex', 'Embarked']:
    le = LabelEncoder()
    df1[col] = le.fit_transform(df1[col])
    label_encoders[col] = le
```

```python
df1.drop(columns=['Name', 'Ticket', 'PassengerId'], inplace=True)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import xgboost as xgb
import lightgbm as lgb

# 1. Split dataset
x = df1.drop(columns='Survived')
```

```python
x
```

```python
y = df['Survived']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
lgb_model = lgb.LGBMClassifier(random_state=42)
y_pred=lgb_model.fit(x_train, y_train)
y_pred
```

```python
y_pred_lgb = lgb_model.predict(x_test)
y_pred_lgb
```

```python
lgb_accuracy = accuracy_score(y_test, y_pred_lgb)
print(lgb_accuracy)
lgb_precision = precision_score(y_test, y_pred_lgb)
print(lgb_precision)
lgb_recall = recall_score(y_test, y_pred_lgb)
print(lgb_recall)
lgb_f1 = f1_score(y_test, y_pred_lgb)
print(lgb_f1)
print( classification_report(y_test, y_pred_lgb))
```

```python
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(x_train, y_train)
```

```python
y_pred_xgb = xgb_model.predict(x_test)
y_pred_xgb
```

```python
# Comparative Analysis:
xgb_accuracy = accuracy_score(y_test, y_pred_xgb)
print(xgb_accuracy)
xgb_precision = precision_score(y_test, y_pred_xgb)
print(xgb_precision)
xgb_recall = recall_score(y_test, y_pred_xgb)
print(xgb_recall)
xgb_f1 = f1_score(y_test, y_pred_xgb)
print(xgb_f1)
print( classification_report(y_test, y_pred_xgb))
```

```python
import pandas as pd
comparison_df = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'XGBoost': [xgb_accuracy, xgb_precision, xgb_recall, xgb_f1],
    'LightGBM': [lgb_accuracy, lgb_precision, lgb_recall, lgb_f1]
})
comparison_df.set_index('Metric', inplace=True)
```

```python
import matplotlib.pyplot as plt
comparison_df.plot(kind='bar', figsize=(10, 6), colormap='viridis')
plt.title('Model Performance Comparison: XGBoost vs LightGBM')
plt.ylabel('Score')
plt.ylim(0, 1)
plt.grid(True, axis='y')
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
